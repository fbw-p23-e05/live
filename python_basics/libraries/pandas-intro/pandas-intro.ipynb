# Cleaning data with Pandas
1. Dealing with missing data by dropping missing values.
2. Dealing with missing data by replacing the values.
3. Dealing with duplicate data
4. Renaming columns

## Dropping missing values
- We use `dropna()` method to drop missing values.
- Losing a small portion of data impact the analysis

```python
df2 = df.copy()
```

```python
df2.dropna()
```

## Replacing the missing Values
- fillna() method to replace missing values.

```python
df3 = df.copy()
```

```python
df3.fillna("None")
```

## Duplicate data
- `drop_duplicates()` method
- Drops 1 or more duplicate of the same record(row).

```python
df4 = pd.read_csv("country_full.csv")
```

```python
df4
```

```python
df4.drop_duplicates()
```

## Renaming Columns

- `rename()`
- key-value pair - `{"old_name": "new_name"}`

```python
df5 = df.copy()
```

```python
df5.rename(columns={"iso_3166-2": "ISO_3166-2"})
```

## Summary operators for Data Analysis
- Mean(Average)
- Mode(Most Frequent)
- Median(Middle number)

```python
df["region-code"].mean()
```

```python
df["region-code"].mode()
```

```python
df["region-code"].median()
```
